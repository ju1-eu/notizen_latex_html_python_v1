```python
import datetime

def add_or_replace_comment(content, filename):
    # Erstellen Sie einen Zeitstempel für heute
    TIMESTAMP = datetime.datetime.now().strftime('%d-%b-%y')
    
    # Der Kommentar, der hinzugefügt oder ersetzt werden soll
    new_comment = f"% ju {TIMESTAMP} {filename}"

    # Entfernen Sie alle Zeilen, die mit "% ju" beginnen
    content_lines = [line for line in content.splitlines() if not line.strip().startswith("% ju")]

    # Fügen Sie den neuen Kommentar am Anfang der aktualisierten Zeilenliste hinzu
    content_lines = [new_comment] + content_lines

    # Geben Sie den zusammengesetzten Inhalt zurück
    return "\n".join(content_lines)

# Beispiel
filename = "01-Grundbegriffe-Hinweise-Signale.tex"
content = '''% ju 9-Sep-23 01-Grundbegriffe-Hinweise-Signale.tex
% ju 8-Sep-23 01-Grundbegriffe-Hinweise-Signale.tex'''
content = add_or_replace_comment(content, filename)
print(content)
```

```python
import re

# Dein LaTeX-Text mit den longtable-Umgebungen
latex_text = """
\\begin{longtable}[]{@{}cc@{}}
\\toprule
\\textbf{Standard-Sicherung} & $R_\\text{Sicherung}$ \\
\\midrule
\\endhead
5 A & 15 $\\text{m} \\Omega$ \\
10 A & 7,5 $\\text{m} \\Omega$ \\
\\bottomrule
\\end{longtable}

\\begin{longtable}[]{@{}cc@{}}
\\toprule
\\textbf{Mini-Sicherung} & $R_\\text{Sicherung}$ \\
\\midrule
\\endhead
5 A & 16,7 $\\text{m} \\Omega$ \\
7,5 A & 10 $\\text{m} \\Omega$ \\
10 A & 7 $\\text{m} \\Omega$ \\
\\bottomrule
\\end{longtable}
"""

# Suche nach longtable-Umgebungen und entferne \endhead
search_pattern = r'\\begin{longtable}(.+?)\\end{longtable}'
replacement_pattern = r'\\begin{table}[!ht]\n\\centering\n\\caption{}%\n\\begin{tabular}\1\\end{table}'

# Führe die Suche und Ersetzung durch, dabei \endhead entfernen
result_text = re.sub(search_pattern, replacement_pattern, latex_text, flags=re.DOTALL)

# Entferne \endhead und Leerzeilen aus dem Ergebnis
result_text = result_text.replace('\\endhead', '').strip()

# Gib das Ergebnis aus
print(result_text)

```

```python
import re

def replace_tables(match):
    column_spec = match.group(1)
    table_content = match.group(2)

    # Entferne doppelte \\toprule
    table_content = re.sub(r"\\\\toprule\s+\\\\toprule", r"\\\\toprule", table_content)
    
    # Entferne \\endhead
    table_content = table_content.replace("\\endhead", "")
    
    # Erstelle die neue table-Umgebung
    return f"""\\begin{{table}}[!ht]
\\centering
\\caption{{}}% \\label{{tab:}}%% anpassen
\\begin{{tabular}}{{@{{}}{column_spec}@{{}}}}
{table_content}
\\end{{tabular}}
\\end{{table}}"""

# Testdaten
content = """
\\begin{longtable}[]{@{}lll@{}}
\\toprule
\\textbf{Spalte 1} & \\textbf{Spalte 2} & \\textbf{Spalte 3} \\
\\midrule
\\endhead
Zelle 1 & Zelle 2 & Zelle 3 \\
Zelle 4 & Zelle 5 & Zelle 6 \\
Zelle 7 & Zelle 8 & Zelle 9 \\
\\bottomrule
\\end{longtable}
"""

# Ersetzen Sie alle longtable-Umgebungen und deren Inhalt durch table
new_content = re.sub(r'\\begin{longtable}\[\]{@{}(.*?)@{}}(.+?)\\end{longtable}', replace_tables, content, flags=re.DOTALL)
print(new_content)
```
